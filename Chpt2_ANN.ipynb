## Chapter 2: Artificial Neural Networks

### 2.1: The main element of Artificial Neural Networks: The Perceptron


A Neural Network (NN) or Artificial Neural Network (ANN) is a collection of so-called `perceptrons`. Perceptrons are the main building blocks of ANNs. Perceptrons take numerical inputs, combine these inputs based on some logical form (= perceive), activate these inputs according to certain rules and then transform the resulting output into subsequent actions, or outputs.

Perceptrons are interconnected with other perceptrons within a layer and between layers. Consequently, the actions transformed from inputs through the perceptron are then transmitted to other, interconnected perceptrons, which use the actions, or outputs, themselves as inputs and perform the same transformative operation as before, thereby enabling a flow in which inputs are transformed and sent as actions through an artificial network until the desired output is achieved.

ANNs are derived from Biological Neuron Networks and thus their elements and structure share strong similarities with the main blueprint of the sensory or nervous system apparent within biological brains. Relating to our biological system, neurons are the fundamental cells of the brain. They receive sensory input from external sources, activate these inputs and transform them into electrical signals which then send motor commands to our muscles, leading to subsequent muscular functions which we define as either reflexes or cognitive actions.

The biological process of taking an electrical signal and transforming it into a muscular action is executed in a so-called `Neuron Function`. Perceptrons are essentially derived from neuron functions and their structure models the sensory transmission evident in our biological systems.



### 2.2 The key parameters of Perceptrons

In order to mimic the biological systems, any perceptron is a neuron function consisting of four key parameters:

- x: inputs
- $w$: weights
- $\beta$:  bias term
- $\lambda$: activation function

These four key parameters are associated according to the following functional relationship:

-------------------------------------------------------------
**Definition 1**:

A neuron function $f: \mathbb R^n \rightarrow \mathbb R$ is a mapping of the form:

$$f(x) = \lambda(w \cdot x + \beta)$$

where we have:

- $\lambda : \mathbb R^n \rightarrow \mathbb R$: Continuous non-linear function called `activation function`
- $w \in \mathbb R$: Vector of `weights`
- $\beta$: `Bias` term

------------------------------------------------------------------------------

Essentially, what a perceptron (or neuron function) does is as follows:

1. It takes individual inputs
2. These inputs are weighted with their corresponding weights
3. A bias term is added
4. The resulting numerical value is transformed through a continuous, non-linear function

Based on the output generated by 4, the resulting perceptron is either activated and transmits the output through to interconnected perceptrons throughout the network or remains deactivated, thereby stopping the transmission.

Let's explore the individual elements a little more in detail

#### 2.2.1: The linear function

The first operation is associated with the inputs and their corresponding weights $w$. As we can see by the dot-product ($\cdot$), we perform an element-wise multiplication and take the sum product of the elements. We then add a bias term which can be thought of as threshold the perceptron must reach before the output is generated. This results in a linear function for each perceptron.

#### 2.2.2: The activation function

As we understand, the first step transforms the input into a linear function. Traditional Machine Learning (ML) algorithms are usually based on linearity assumptions, thereby including said assumptions when deriving an equation. However, given that most relationships are indeed non-linear, linear transformations alone cannot fully capture these relationships. As a consequence, we need a component which induces non-linearity from the data. This is achieved through an activation function.

There are multiple activation functions proposed by the literature. To give you an overview:



image.png

Although multiple activation functions with different properties exist, the main ones used are:

- ReLU
- Tanh
- Sigmoid
- Softmax


A typical example of an activation function is called `ReLU` (Rectified Linear Unit). It is defined as:

$$
ReLU(x) = \begin{cases}
    x,& \text{if } x > 0\\
    0,              & \text{otherwise}
\end{cases}
$$

ReLU is a simple non-linear transformation which models thes threshold between two states: active and inactive. Consequently, it is a simple way to model neurons in a biological system with two states: Resting (inactive) and Firing (active). The main idea is that impulses transmitted from other neurons can change the state from resting to active if the impulses reach the desired threshold. As such, the change of the ReLU function from 0 to the value of x provided is able to reflect the transmission from an inactive to an active state, thereby mimicking the biological system in which a neuron fires an electrical impulse through the network if being activated.

### 2.3: Feed-Forward Neural Networks

#### 2.3.1: Layers

The behaviour of a particular perceptron is mimicked by multiple other perceptrons. Multiple perceptrons are interconnected with each other. These interconnected perceptrons form a so-called `layer` (or layer function) of the following form:

-------------------------------------------------------------
**Definition 2**:

A layer function $g: \mathbb R^n \rightarrow \mathbb R^m$ is a mapping of the form:

$$g(x) = [f_1(x), f_2(x),...,f_m(x)] = [\lambda(w_1^1 \cdot x_1^0 + \beta_1^1), \lambda(w_2^1 \cdot x_1^0 + \beta_2^1), \lambda(w_m^1 \cdot x_m^0 + \beta_m^1)]$$

where each $f_i$ is a perceptron with its:
- own vector of weights $w_i = (w_{i1},..., w_{in})$
- own scalar of biases $\beta_i, i = 1, ..., m$

------------------------------------------------------------------------------

Given the vector structure of the input parameters, weights and biases, we understand that the layer function is defined as a matrix of paramters:

$$
W =
\begin{pmatrix}
w_{11} &\dots & w_{1n} \\
\vdots & \ddots & \vdots\\
w_{m1} & \dots & w_{mn}  \\
\end{pmatrix}
$$

and a vector of biases:

$$
\beta =  
\begin{pmatrix}
\beta_{1} \\
\vdots \\
\beta_{m} \\
\end{pmatrix}
$$

and it can be written as:

$$
g(x) = \tilde \lambda (Wx + \beta)
$$

In essence, we have three distinct forms of layers:

1. Input Layers:
  - Input Layers are the first set of layers within an ANN
  - They serve as the usual input parameters we feed into the model
2. Hidden Layers:
  - Hidden Layers are the intermediate layers
  - They take the input parameters and transmit then throughout the network
  - They are the main layers for the computational operations within a network
3. Output Layers:
  - Output layers are the last set of layers within the ANN
  - They aggregate the computed results and produce the final output



#### 2.3.2: Multilayer Perceptron

Through interconnecting multiple perceptrons, we generated a layer. We can expand the same logic throughout multiple, fully interconnected layers to form blocks or multi-layered architectures. Through "layering" multiple layers and connecting their individual perceptrons with one another, we can finally create what is called a `Multilayer Perceptron` (MLP). MLPs are a synonym for the typical feedforward ANNs, consisting of fully connected perceptrons (neurons), nonlinear activation functions and a depth of at least three layers.

Mathematically, an ANN can be defined as follows:

-------------------------------------------------------------
**Definition 3**:

An Artificial Neural Network (ANN) (=MLP) $h: \mathbb R^n \rightarrow \mathbb R^m$ is a function of the form:

$$h(x) = h_M \circ h_{M-1} \circ \dots \circ h_1(x), M \geq 1$$

where $h_i$ is a `layer` function with its own matrix of parameters $w^i$ and vector $\beta^i$.

As such, we can expand the ANN function into its layer functions:

$$h(x) = [f_1^1(x), f_2^1(x),...,f_m^1(x)] \circ  \dots \circ [f_1^{M-1}(x), f_2^{M-1}(x),...,f_m^{M-1}(x)] \circ [f_1^{M}(x), f_2^{M}(x),...,f_m^{M}(x)]$$ \\

We can further clarify the relationship by expanding it into its perceptrons:

$$h(x) = [\lambda(w_{1,1}^1 \cdot x_1^0 + w_{1,2}^1 \cdot x_2^0 + \dots + w_{1,m}^1 \cdot x_m^0 + \beta_1^1), \lambda(w_{2,1}^1 \cdot x_1^0 + w_{2,2}^1 \cdot x_2^0 + \dots + w_{1^2,m}^1 \cdot x_m^0 + \beta_1^1), \lambda(w_{m,1}^1 \cdot x_1^0 + w_{m,2}^1 \cdot x_2^0 + \dots + w_{m,m}^1 \cdot x_m^0 + \beta_1^1)], [\lambda(w_{1,1}^2 \cdot x_1^1 + w_{1,2}^2 \cdot x_2^1 + \dots + w_{1,m}^2 \cdot x_m^1 + \beta_1^2), \lambda(w_{2,1}^2 \cdot x_1^1 + w_{2,2}^2 \cdot x_2^1 + \dots + w_{1^2,m}^2 \cdot x_m^1 + \beta_1^2), \lambda(w_{m,1}^2 \cdot x_1^1 + w_{m,2}^2 \cdot x_2^1 + \dots + w_{m,m}^2 \cdot x_m^1 + \beta_1^2)], \dots, [\lambda(w_{1,1}^m \cdot x_1^{n-1} + w_{1,2}^n \cdot x_2^{n-1} + \dots + w_{1,m}^n \cdot x_m^{n-1} + \beta_1^n), \lambda(w_{2,1}^n \cdot x_1^{n-1} + w_{2,2}^n \cdot x_2^n + \dots + w_{1^2,m}^n \cdot x_m^{n-1} + \beta_1^n), \lambda(w_{m,1}^n \cdot x_1^{n-1} + w_{m,2}^n \cdot x_2^{n-1} + \dots + w_{m,m}^n \cdot x_m^{n-1} + \beta_1^n)]$$

where:
- $w_i^j = (w_{i1}^j,..., w_{im}^j)$
- $\beta_i^j, i = 1, ..., m, i = 1,...,M$
------------------------------------------------------------------------------

Let's now dive into the specific layers of an ANN. Note that we are only going to discuss hidden and output layers as the input layer is usually only defined as the input vector of data, $x$.

##### 2.3.2.1: Hidden Layers - The main Network Architecture of an ANN

Let's visualise the formula defined above with a simple example. Let's draw an ANN with two layers. Both layers are defined as `hidden layers`. As such, the ANN only consists of the following form:

$$h(x) = [\lambda(w_{1,1}^1 \cdot x_1^0 + w_{1,2}^1 \cdot x_2^0 + \dots + w_{1,m}^1 \cdot x_m^0 + \beta_1^1), \\ \lambda(w_{2,1}^1 \cdot x_1^0 + w_{2,2}^1 \cdot x_2^0 + \dots + w_{1^2,m}^1 \cdot x_m^0 + \beta_1^1), \\ \lambda(w_{m,1}^1 \cdot x_1^0 + w_{m,2}^1 \cdot x_2^0 + \dots + w_{m,m}^1 \cdot x_m^0 + \beta_1^1)]
$$

image.png

Here, we show you the visual architecture and the mathematical notation of a `fully-connected feed forward artificial neural network`.

The ANN is called `fully connected` because for each *except the last layer* (=output layer) each perceptron is connected with all **perceptrons of the previous and the subsequent layer**. Consequently, each node in each layer is connected to each node in the two adjacent layers.

With regards to the mathematical notation, we understand the following structure:

- $w_{i,m}^n$
  - n: The layer position within the network of the current perceptrons
  - i: The position of the respective input perceptron at the current layer
  - m: The position of the respective output perceptron at the previous layer


##### 2.3.2.2: Output Layers - ANN classifiers and the softmax function

The output layer of an ANN is usually defined throughout a `soft classifier function`. By definition, a soft classifier creates class probabilities. As such, it provides an output of list of probabilities that output $y$, generated through the fully-connected network with input $x$, belongs to class $c$. For instance, if we have images of cats and dogs, the soft classifier function takes the transformed inputs and assigns a probability of i to dogs and a probability of (1-i) to cats.

More generally, we can state that if $\phi(x) = (\phi_1(x), \dots, \phi_m(x)$ is a soft classifier, we need that for each $x$:

$$
0 \leq \phi_i(x) \leq 1, i = 1, \dots, m \ni \sum_{i=1}^m \phi_i(x) = 1
$$

This means that a soft classifier must provide a probability of output $y$ to each class such that:

- Each class has a probability between 0 and 1
- The sum of all class probabilities sums up to 1

Based on this, we can formally define the **ANN classifier in the output layer**:

-------------------------------------------------------------
**Definition 4**:

An ANN `classifier` is a function $\tilde h: \mathbb R^n \rightarrow \mathbb R^m$ of the form:

$$
\tilde h(x) = \sigma \circ h(x)
$$

where $\tilde h: \mathbb R^n \rightarrow \mathbb R^m$ is an ANN and $\sigma: \mathbb R^n \rightarrow \mathbb R^m$ is the `softmax` function, defined as:

$$
\sigma(y) = (\frac{e^{y_1}}{\sum_{i=1}^m e^{y_1}}, \dots, \frac{e^{y_m}}{\sum_{i=1}^m e^{y_i}})
$$


--------------------------------------------------------------------------

As one can understand, the softmax function normalises the output of $\tilde h$ such that the two requirements stated above are satisfied. Moreover, the softmax function is **order preserving** as $y \rightarrow e^y$ is also monotonically increasing. Consequently, if $y_j > y_i$, then $ \sigma(y)_i > \sigma(y)_j$, providing that the largest component of $h(x)$ is also the one with the highest class probability $\tilde h(x)$.

### 2.4: The universal approximation thereom


We understand that any ANN is a function created through an **iterative composition of affine functions and non-linear transformations**.

Remember that an **affine** function is nothing else than a linear transformation followed by a translation, mathematically described as

$$
f(X) = A \cdot x + b
$$

A non-linear transformation then is simply a mathematical operation that creates a function which does not adhere to the properties of linearity, thereby capturing more complex patterns and relationships in data that cannot be effectively represented using linear functions.



Based on this, we can look at the *second* definition of the `Universal Approximation Theorem`:

-------------------------------------------------------------
**Definition 5**:

Let $K \subset \mathbb R^n$ be a closed and bounded set and let $\lambda: \mathbb R \rightarrow \mathbb R$ be a non-linear activation function which is **not a polynomial**. Let $f: K \rightarrow \mathbb R^n$ be a continuous function. Then, for any $\epsilon > 0$ there exists $k \in \mathbb N$ and affine functions $L_1: \mathbb R^n \rightarrow \mathbb R^k$ and $L_2: \mathbb R^k \rightarrow \mathbb R^m$ such that:

$$
\sup_{x \in K}||f(x) - L_2 \circ \tilde \lambda \circ L_1(x)|| < \epsilon
$$

--------------------------------------------------------------------------

To put the formula into words (and define some notation on how to state a proper approximation), we can say the following:

The Universal Approximation Theorem, as expressed through the given function, states that for:

1. any continuous function $f(x)$ defined on a compact set K,
2. there exists a neural network composed of two affine transformations followed by a non-linear activation function
  - represented by $L_1, L_2, and \lambda$, respectively,
3. such that the supremum (maximum) norm of the difference between $f(x)$ and the composition of $L_2, \lambda, and L_1$ applied to $x$
  - so the iterative connection of affine and non-linear transformations within the network
4. is less than a given epsilon $\epsilon$,
5. indicating that the neural network can approximate the function f(x) within the desired accuracy.

To put it bluntly, it states that a feedforward neural network with k hidden layers, where each hidden layer implements an affine transformation followed by a non-linear activation function, can **approximate any continuous function to any desired level of accuracy**, provided the activation functions meet certain condition.

#### 2.4.1: Polynomial Activation functions

The reason why we cannot have a polynomial activation function is the following. Let's assume that $\lambda$ is a polynomial of degree d. Let's also assume that $L_1$ is an affine map with k = 1 hidden layer and $L_2$ is also an affine map with. Then, we understand that $L_2 \circ \tilde \lambda  L_1$ is a linear combination of perceptrons $\tilde \lambda(a_1x + b_1)$. This provides us with two realisations:

1. A polynomial of degree d composed with an affine function is still a polynomial of degree d
2. A linear combination of polynomials of degree d is a polynomial of degree d or less

Recall that the Weierstrass approximation theorem states that arbitrarily close approximations can be achieved by polynomial functions, and in general, an arbitrarily large degree may be needed.

Consequently, $\tilde \lambda(a_1x + b_1)$ is a polynomial of degree degree $ \leq d$ and can at thus just best possibly approximate a function which can be approximated by d-degree polynomials.
