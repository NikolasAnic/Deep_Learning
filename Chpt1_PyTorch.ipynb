### 1.1: Introduction to Tensors

Tensors are a fundamental concept in various fields such as mathematics, physics, and especially in machine learning and deep learning. At its core, a tensor is a mathematical object that generalizes the concepts of scalars, vectors, and matrices to higher dimensions. Tensors can be thought of as multi-dimensional arrays or containers that hold numerical values.

In the context of deep learning and machine learning, tensors are used to represent data, including inputs, outputs, and intermediate representations in neural networks.

Tensors can have different ranks, which correspond to the number of dimensions they have:

 - Scalar (Rank 0 Tensor): A single number, such as a floating-point number or an integer

- Vector (Rank 1 Tensor): An array of numbers along a single dimension

- Matrix (Rank 2 Tensor): An array of numbers arranged in rows and columns

- Higher-Dimensional Tensors (Rank > 2): Tensors with more than two dimensions, such as a 3D tensor or a tensor with even higher dimensions

Let's see this as an example:





```python
import torch

class Tensor:
    def __init__(self, data):
        self.data = torch.tensor(data)
        self.shape = self.data.shape
        self.rank = len(self.shape)
        self.ndim = self.data.ndim

# Create scalar, vector, matrix, and tensor objects
scalar_tensor = Tensor(5)
vector_tensor = Tensor([1, 2, 3])
matrix_tensor = Tensor([[1, 2, 3], [4, 5, 6]])
tensor_3d = Tensor([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

print('-'*50)
print("Scalar Tensor:", scalar_tensor.data, " Shape:", scalar_tensor.shape, ' Number of Dimensions:', scalar_tensor.ndim, ' Rank:', scalar_tensor.rank)
print('-'*50)
print("Vector Tensor:", vector_tensor.data, " Shape:", vector_tensor.shape, ' Number of Dimensions:', vector_tensor.ndim, ' Rank:', vector_tensor.rank)
print('-'*50)
print("Matrix Tensor:", matrix_tensor.data, " Shape:", matrix_tensor.shape, ' Number of Dimensions:', matrix_tensor.ndim, ' Rank:', matrix_tensor.rank)
print('-'*50)
print("3D Tensor:", tensor_3d.data, " Shape:", tensor_3d.shape, ' Number of Dimensions:', tensor_3d.ndim, ' Rank:', tensor_3d.rank)
print('-'*50)
```

#### 1.1.1: Random Tensors

Random tensors are tensors whose values are generated using random processes. They are commonly used in machine learning and deep learning for various purposes, such as initializing weights in neural networks, generating noise, or creating random samples for statistical analysis.

In the context of torch, you can create random tensors using the torch.rand function. This function generates random values between 0 and 1, following a uniform distribution. You can also use other functions like torch.randn to generate random values from a standard normal distribution (mean 0, standard deviation 1).

```python
import torch

class Tensor:
    def __init__(self, data=None, shape=None, random=False):
        if random:
            self.data = torch.rand(shape) if shape is not None else torch.rand(1)
        else:
            self.data = torch.tensor(data) if data is not None else None
        self.shape = self.data.shape if self.data is not None else shape
        self.rank = len(self.shape) if self.shape is not None else None
        self.ndim = self.data.ndim if data is not None else None

# Create a random scalar tensor
scalar_tensor = Tensor(random=True)
# Create a random vector tensor
vector_tensor = Tensor(shape=(3,), random=True)
# Create a random matrix tensor
matrix_tensor = Tensor(shape=(2, 3), random=True)
# Create a random 3D tensor
tensor_3d = Tensor(shape=(2, 2, 2), random=True)

print('-'*50)
print("Scalar Tensor:", scalar_tensor.data, " Shape:", scalar_tensor.shape, ' Number of Dimensions:', scalar_tensor.ndim, ' Rank:', scalar_tensor.rank)
print('-'*50)
print("Vector Tensor:", vector_tensor.data, " Shape:", vector_tensor.shape, ' Number of Dimensions:', vector_tensor.ndim, ' Rank:', vector_tensor.rank)
print('-'*50)
print("Matrix Tensor:", matrix_tensor.data, " Shape:", matrix_tensor.shape, ' Number of Dimensions:', matrix_tensor.ndim, ' Rank:', matrix_tensor.rank)
print('-'*50)
print("3D Tensor:", tensor_3d.data, " Shape:", tensor_3d.shape, ' Number of Dimensions:', tensor_3d.ndim, ' Rank:', tensor_3d.rank)
print('-'*50)
```

#### 1.1.2: Zeros and Ones Tensors


Zeros and ones tensors are special types of tensors where all the elements are set to either 0 or 1, respectively. These tensors are commonly used for initialization, padding, and other operations in machine learning and deep learning.

In torch, you can create zeros and ones tensors using the torch.zeros and torch.ones functions, respectively. You provide the desired shape as an argument, and the function returns a tensor with all elements set to the specified value (0 for zeros and 1 for ones).

```python
import torch

class Tensor:
    def __init__(self, data=None, shape=None, random=False, zeros=False, ones=False):
        if random:
            self.data = torch.rand(shape) if shape is not None else torch.rand(1)
        elif zeros:
            self.data = torch.zeros(shape) if shape is not None else torch.zeros(1)
        elif ones:
            self.data = torch.ones(shape) if shape is not None else torch.ones(1)
        else:
            self.data = torch.tensor(data) if data is not None else None
        self.shape = self.data.shape if self.data is not None else shape
        self.rank = len(self.shape) if self.shape is not None else None
        self.ndim = self.data.ndim if self.shape is not None else None

# Create a zeros scalar tensor
zeros_scalar_tensor = Tensor(zeros=True)
# Create a ones scalar tensor
ones_scalar_tensor = Tensor(ones=True)
# Create a zeros vector tensor
zeros_vector_tensor = Tensor(shape=(3,), zeros=True)
# Create a ones vector tensor
ones_vector_tensor = Tensor(shape=(3,), ones=True)
# Create a zeros matrix tensor
zeros_matrix_tensor = Tensor(shape=(2, 3), zeros=True)
# Create a ones matrix tensor
ones_matrix_tensor = Tensor(shape=(2, 3), ones=True)
# Create a zeros 3D tensor
zeros_3d_tensor = Tensor(shape=(2, 2, 2), zeros=True)
# Create a ones 3D tensor
ones_3d_tensor = Tensor(shape=(2, 2, 2), ones=True)


print('-'*50)
print("Scalar Tensor:", zeros_scalar_tensor.data, " Shape:", zeros_scalar_tensor.shape, ' Number of Dimensions:', zeros_scalar_tensor.ndim, ' Rank:', zeros_scalar_tensor.rank)
print('-'*50)
print("Vector Tensor:", ones_vector_tensor.data, " Shape:", ones_vector_tensor.shape, ' Number of Dimensions:', ones_vector_tensor.ndim, ' Rank:', ones_vector_tensor.rank)
print('-'*50)
print("Matrix Tensor:", zeros_matrix_tensor.data, " Shape:", zeros_matrix_tensor.shape, ' Number of Dimensions:', zeros_matrix_tensor.ndim, ' Rank:', zeros_matrix_tensor.rank)
print('-'*50)
print("3D Tensor:", ones_3d_tensor.data, " Shape:", ones_3d_tensor.shape, ' Number of Dimensions:', ones_3d_tensor.ndim, ' Rank:', ones_3d_tensor.rank)
print('-'*50)
```

#### 1.1.3: Arange functions

The torch.arange function is used to generate a 1-D tensor with values ranging from a specified start value up to, but not including, a specified end value. You can also specify the step value, which determines the spacing between the generated values.

```python
import torch

class Tensor:
    def __init__(self, data=None, shape=None, random=False, zeros=False, ones=False, arange=None):
        if random:
            self.data = torch.rand(shape) if shape is not None else torch.rand(1)
        elif zeros:
            self.data = torch.zeros(shape) if shape is not None else torch.zeros(1)
        elif ones:
            self.data = torch.ones(shape) if shape is not None else torch.ones(1)
        elif arange is not None:
            self.data = torch.arange(*arange)
        else:
            self.data = torch.tensor(data) if data is not None else None
        self.shape = self.data.shape if self.data is not None else shape
        self.rank = len(self.shape) if self.shape is not None else None
        self.ndim = self.data.ndim if self.shape is not None else None

# Create a range tensor from 0 to 4
range_tensor = Tensor(arange=(4,))
# Create a range tensor from 2 to 10 with step 2
range_with_step_tensor = Tensor(arange=(2, 10, 2))

print('-'*50)
print("Scalar Tensor:", range_tensor.data, " Shape:", range_tensor.shape, ' Number of Dimensions:', range_tensor.ndim, ' Rank:', range_tensor.rank)
print('-'*50)
print("Vector Tensor:", range_with_step_tensor.data, " Shape:", range_with_step_tensor.shape, ' Number of Dimensions:', range_with_step_tensor.ndim, ' Rank:', range_with_step_tensor.rank)
```

#### 1.1.4: Tensor datatypes

The default datatype of a Tensor in PyTorch is `32-bit floating point`. However, PyTorch offers many different options of the datatypes. To obtain a full list on the datatypes, you can consult [the tensor datatypes userguide of PyTorch](https://pytorch.org/docs/stable/tensors.html#data-types).

Note that **non-matching datatypes of tensors** are one of the **most common error reasons** in PyTorch. As such, you should always ensure that the datatypes are equal throughout all the Tensors you are using.

Let's provide you with a small example to see the different datatypes.

```python
import torch

class DataTypePrinter:
    def __init__(self, data_list):
        self.data_list = data_list
        self.create_tensors()

    def create_tensors(self):
        self.default_tensor = torch.tensor(self.data_list)
        self.float16_tensor = torch.tensor(self.data_list, dtype=torch.float16)
        self.float64_tensor = torch.tensor(self.data_list, dtype=torch.float64)

    def print_tensors(self):
        print("Default (float32) tensor:")
        print(self.default_tensor)
        print("\nFloat16 tensor:")
        print(self.float16_tensor)
        print("\nFloat64 tensor:")
        print(self.float64_tensor)

# Given data list
data_list = [1.0, 2.0, 3.0, 4.0]

# Create a DataTypePrinter object and print the tensors
printer = DataTypePrinter(data_list)
printer.print_tensors()
```

#### 1.1.5: Retrieve information from tensors

There are mainly three sources of information you can retrieve from tensors:

- `shape`: The dimensions of the tensor
- `ndim`: The number of dimensions of the tensor
- `rank`: The rank of the tensor
- `dtype`: The datatype of the tensor
- `device`: The device the tensor is stored on (CPU vs. GPU)

Getting back to our earlier example code, we can print these three values:

```python
import torch

class Tensor:
    def __init__(self, data=None, shape=None, random=False, zeros=False, ones=False, arange=None):
        if random:
            self.data = torch.rand(shape) if shape is not None else torch.rand(1)
        elif zeros:
            self.data = torch.zeros(shape) if shape is not None else torch.zeros(1)
        elif ones:
            self.data = torch.ones(shape) if shape is not None else torch.ones(1)
        elif arange is not None:
            self.data = torch.arange(*arange)
        else:
            self.data = torch.tensor(data) if data is not None else None
        self.shape = self.data.shape if self.data is not None else shape
        self.rank = len(self.shape) if self.shape is not None else None
        self.ndim = self.data.ndim if self.shape is not None else None
        self.dtype = self.data.dtype if self.shape is not None else None
        self.device = self.data.device if self.shape is not None else None

# Create a range tensor from 0 to 4
range_tensor = Tensor(arange=(4,))

print('-'*130)
print("Scalar Tensor:", range_tensor.data, " Shape:", range_tensor.shape, ' Number of Dimensions:', range_tensor.ndim, ' Rank:', range_tensor.rank, ' Datatype:', range_tensor.dtype, ' Device:', range_tensor.device)
print('-'*130)
```

#### 1.1.6: Tensor operations

Any Deep Learning model learns by performing iterative combinations of the following six operations on its tensors:

- Addition
- Subtraction
- Multiplication
- Division
- Matrix Multiplication
- Comparison

These are the main building blocks of PyTorch operations. Let's quickly introduce them:

```python
import torch

class TensorOperations:
    def __init__(self, tensor1, tensor2):
        self.tensor1 = tensor1
        self.tensor2 = tensor2

    def addition(self):
        return self.tensor1 + self.tensor2

    def subtraction(self):
        return self.tensor1 - self.tensor2

    def multiplication(self):
        return self.tensor1 * self.tensor2

    def division(self):
        return self.tensor1 / self.tensor2

    def matrix_multiplication(self):
        return torch.matmul(self.tensor1, self.tensor2)

    def comparison(self):
        return self.tensor1 > self.tensor2

# Create tensors
tensor1 = torch.tensor([[1, 2], [3, 4]])
tensor2 = torch.tensor([[5, 6], [7, 8]])

# Create a TensorOperations object
tensor_ops = TensorOperations(tensor1, tensor2)

# Perform operations and print results
print("-" * 40)
print("Tensor 1:\n", tensor_ops.tensor1)
print("-" * 40)
print("Tensor 2:\n", tensor_ops.tensor2)
print("-" * 40)
print("Addition Result:\n", tensor_ops.addition())
print("-" * 40)
print("Subtraction Result:\n", tensor_ops.subtraction())
print("-" * 40)
print("Multiplication Result:\n", tensor_ops.multiplication())
print("-" * 40)
print("Division Result:\n", tensor_ops.division())
print("-" * 40)
print("Matrix Multiplication Result:\n", tensor_ops.matrix_multiplication())
print("-" * 40)
print("Comparison Result:\n", tensor_ops.comparison())
print("-" * 40)
```

#### 1.1.7: Common aggregation of tensors

We can also aggregate tensors. The aggregation commands are largely identical to what we know from Numpy, so it's easy to replicate them. Among the most common ones are:

- Sum
- Mean
- Min
- Max
- Median
- Argmin
- Argmax
- Variance

Importantly, for the Mean and Variance, we need to change the datatype to the default float of 32 bits.

```python
import torch

class AggregationOperations:
    def __init__(self, tensor):
        self.tensor = tensor

    def calculate_min(self):
        return torch.min(self.tensor)

    def calculate_max(self):
        return torch.max(self.tensor)

    def calculate_sum(self):
        return torch.sum(self.tensor)

    def calculate_mean(self):
        return torch.mean(self.tensor.type(torch.float32))

    def calculate_argmin(self):
        return torch.argmin(self.tensor)

    def calculate_argmax(self):
        return torch.argmax(self.tensor)

    def calculate_median(self):
        return torch.median(self.tensor).item()

    def calculate_variance(self):
        return torch.var(self.tensor.type(torch.float32)).item()

# Create a tensor
tensor = torch.tensor([5, 2, 8, 1, 4, 9, 3, 6])

# Create an AggregationOperations object
aggregation_ops = AggregationOperations(tensor)

# Calculate and print aggregation results
print("Original Tensor:\n", tensor)
print("Minimum Value:", aggregation_ops.calculate_min().item())
print("Maximum Value:", aggregation_ops.calculate_max().item())
print("Sum:", aggregation_ops.calculate_sum().item())
print("Mean:", aggregation_ops.calculate_mean().item())
print("Index of Minimum Value:", aggregation_ops.calculate_argmin().item())
print("Index of Maximum Value:", aggregation_ops.calculate_argmax().item())
print("Median:", aggregation_ops.calculate_median())
print("Variance:", aggregation_ops.calculate_variance())
```

#### 1.1.8: Reshape Operations

PyTorch also provides the same reshaping and data dimension manipulation techniques as Numpy. The most prominent uses are in the following table:

| Method | One-line description |
|--------|----------------------|
| [`torch.reshape(input, shape)`](https://pytorch.org/docs/stable/generated/torch.reshape.html#torch.reshape) | Reshapes `input` to `shape` (if compatible), can also use `torch.Tensor.reshape()`. |
| [`torch.Tensor.view(shape)`](https://pytorch.org/docs/stable/generated/torch.Tensor.view.html) | Returns a view of the original tensor in a different `shape` but shares the same data as the original tensor. |
| [`torch.stack(tensors, dim=0)`](https://pytorch.org/docs/1.9.1/generated/torch.stack.html) | Concatenates a sequence of `tensors` along a new dimension (`dim`), all `tensors` must be the same size. |
| [`torch.squeeze(input)`](https://pytorch.org/docs/stable/generated/torch.squeeze.html) | Squeezes `input` to remove all the dimensions with value `1`. |
| [`torch.unsqueeze(input, dim)`](https://pytorch.org/docs/1.9.1/generated/torch.unsqueeze.html) | Returns `input` with a dimension value of `1` added at `dim`. |
| [`torch.permute(input, dims)`](https://pytorch.org/docs/stable/generated/torch.permute.html) | Returns a *view* of the original `input` with its dimensions permuted (rearranged) to `dims`. |


```python
import torch

class TorchFunctionUsage:
    def __init__(self):
        self.tensor = torch.tensor([1, 2, 3, 4, 5, 6])

    def demonstrate_functions(self):
        tensor_reshape = torch.reshape(self.tensor, (2, 3))
        tensor_reshape_shape = tensor_reshape.shape
        tensor_view = self.tensor.view(2, 3)
        tensor_stack = torch.stack([self.tensor, self.tensor * 2])
        tensor_squeeze = self.tensor.view(2, 3, 1).squeeze()
        tensor_unsqueeze = self.tensor.unsqueeze(dim=1)
        tensor_permute = self.tensor.view(2, 3).permute(1, 0)

        print('-'*40)
        print("Original Tensor:\n", self.tensor, ' Shape:', self.tensor.shape)
        print('-'*40)
        print("Reshaped Tensor:\n", tensor_reshape, ' Shape:', tensor_reshape_shape)
        print('-'*40)
        print("Viewed Tensor:\n", tensor_view)
        print('-'*40)
        print("Stacked Tensors:\n", tensor_stack)
        print('-'*40)
        print("Squeezed Tensor:\n", tensor_squeeze)
        print('-'*40)
        print("Unsqueezed Tensor:\n", tensor_unsqueeze)
        print('-'*40)
        print("Permuted Tensor:\n", tensor_permute)
        print('-'*40)

# Create an instance of the TorchFunctionUsage class
function_usage = TorchFunctionUsage()

# Call the method to demonstrate the functions
function_usage.demonstrate_functions()
```

#### 1.1.9: Working with tensors on a GPU

Graphics Processing Units (GPUs) are specialized hardware designed to perform parallel processing tasks, particularly those related to rendering graphics and handling large-scale mathematical computations. While they were initially developed for graphics rendering, their architecture and parallel processing capabilities have made them invaluable for accelerating various scientific and computational tasks, including deep learning and tensor computations.

GPUs are well-suited for working with tensors and deep learning tasks due to the following reasons:

- Parallel Processing:
  - GPUs consist of thousands of smaller processing cores that can perform calculations simultaneously. This architecture allows them to handle a large number of calculations in parallel, making them highly efficient for tasks involving matrices and tensors, which are fundamental in deep learning.

- Matrix Operations:
  - Deep learning models, especially neural networks, heavily rely on matrix operations, such as matrix multiplications, convolutions, and element-wise operations. GPUs excel at performing these operations in parallel, significantly speeding up the training and inference processes.

- Data Parallelism:
  - Many deep learning algorithms involve the same operations being applied to multiple data points (e.g., images in a batch). GPUs can process these operations in parallel across the data points, leading to faster computation times.

- Large-scale Data:
  - Deep learning often involves working with large datasets. GPUs have memory bandwidths optimized for data throughput, allowing them to efficiently handle large amounts of data without causing performance bottlenecks.

Let's show you how to create a simple model and run it on a GPU configuration. Doing so, we need to put both the input data (the tensors) and the model onto the GPU.

In this example, the `YourModel` class is a simple feedforward neural network with **two fully connected (linear) layers**. It takes an **input tensor of size 10** and **outputs a single value**. The forward method defines the forward pass of the network, where the input tensor is passed through the layers with a **ReLU activation** applied to the first layer.

```python
import torch
import torch.nn as nn
import time

# Create a simple model class
class YourModel(nn.Module):
    def __init__(self):
        super(YourModel, self).__init__()
        self.fc1 = nn.Linear(in_features=10, out_features=5)
        self.fc2 = nn.Linear(in_features=5, out_features=1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# Create an instance of the YourModel class
model = YourModel()

# Generate random input data
input_data = torch.rand(1, 10)

# Run on GPU if available
if torch.cuda.is_available():
    device = torch.device("cuda")
    model.to(device)
    input_data = input_data.to(device)
    print("Running on GPU")
else:
    device = torch.device("cpu")
    print("Running on CPU")

# Measure GPU/CPU inference time
with torch.no_grad():
    start_time = time.time()
    output = model(input_data)
    end_time = time.time()
    inference_time = end_time - start_time

# Print the results
print("Input Data:", input_data)
print("Output:", output)
print("Inference Time:", inference_time, "seconds")
```

Great, we were able to perform Connection to GPU with the following four steps:

- `torch.cuda_is_available()`: Checks if the Cuda option is available and the GPU can be used
- `device = torch.device('cuda')`: Sets option "cuda" to the device of the operation
- `model.to(device)`: Sets the defined model to the device class of "cuda"
- `input_data = input_data.to(device)`: Sets the input data to the device class of "cuda"  

### 1.2: PyTorch Workflows

Now that we introduced you to the main building blocks of PyTorch, it is now time to focus more on the exact workflow of the package. Doing so, we will borrow the content from the great work facilitated by [mrdbourke](https://www.learnpytorch.io/01_pytorch_workflow/) who put up a great introcution into PyTorch workflows and how to use them for Deep Learning applications.

In essence, a PyTorch workflow can be distributed into the following six parts:

| Topic                                   | Contents                                                                                   |
|-----------------------------------------|-------------------------------------------------------------------------------------------|
| **1. Data Preparation**               | This part includes the entire set up of the data preparation. This includes data gathering, sourcing and cleaning, aggregation, and configuration for the model (we skip the additional data analysis part for now) |
| **2. Model Building**                 | This part includes the entire set of the model creation. This includes the choice and set up of the model, the archtitecture and the definition of the hyperparameter space, the choice of the loss functions, optimiser, training loops and weights. |
| **3. Model Fitting** | This part includes the entire training part of the model. This includes the actual feedforward process execution on training data.|
| **4. Model Evaluation** | This part includes the entire set of the model evaluation. This includes the backward propagation and the analysis of the iterative behaviour of the loss function based on both the evalation set (hyperparameter tuning) and the independent test set. |
| **5.Model Saving and Access**        | This part includes the entire set of the model reuse. This includes the saving of the weights and bias terms as well as the continuous access to the model architecture. |

##### 1.2.1.2: Data Splitting

Next, we are going to split the data into training, validation and test set. | Data Type Split                | Purpose                                                                                          | Amount of total data |
|----------------------|--------------------------------------------------------------------------------------------------|----------------------|

Let's separate the data into training, validation and test split.

```python
import torch
from sklearn.model_selection import train_test_split

class DataSplitter:
    def __init__(self, x, y, train_ratio=0.6, val_ratio=0.2, test_ratio=0.2, random_seed=None):
        self.x = x
        self.y = y
        self.train_ratio = train_ratio
        self.val_ratio = val_ratio
        self.test_ratio = test_ratio
        self.random_seed = random_seed

    def split_data(self):
        # Here, we first separate the training and the remaining set. The test set thus is comprised of both the validation and the actual test set
        x_train, x_temp, y_train, y_temp = train_test_split(self.x, self.y, test_size=self.val_ratio+self.test_ratio, random_state=self.random_seed)
        # Calculate the relative validation ratio as fraction of the overall test set ratio
        relative_val_ratio = self.val_ratio / (self.val_ratio + self.test_ratio)
        # Finally, calculate the validation and test set (this is now a 50%-50% split as val_ratio = test_ratio = 0.2)
        x_val, x_test, y_val, y_test = train_test_split(x_temp, y_temp, test_size=relative_val_ratio, random_state=self.random_seed)
        return x_train, y_train, x_val, y_val, x_test, y_test

# Create known parameters
weight_2 = 2.0
weight_1 = -3.0
bias = 1.5

# Create data
start = 0
end = 1
step = 0.02
x = torch.arange(start, end + step, step)
y = weight_2 * x**2 + weight_1 * x + bias

# Instantiate DataSplitter
data_splitter = DataSplitter(x, y, train_ratio=0.7, val_ratio=0.15, test_ratio=0.15, random_seed=42)

# Perform the data split
x_train, y_train, x_val, y_val, x_test, y_test = data_splitter.split_data()

# Print the sizes of the split datasets
print("Train data size:", len(x_train))
print("Validation data size:", len(x_val))
print("Test data size:", len(x_test))
```

#### 1.2.2: Model Building

Let's now start to build our first model to approximate the underlying relationship. We will guide you through the set up of the model step by step.

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Polynomial model class
class PolynomialModel(nn.Module): # The nn.Module: Is a subclass which contains all the building blocks to establish neural networks in PyTorch
    # Here, we initialise the class. Doing so, we need to use the:
      # - `self` argument
      # - `degree` argument: This will provide our model with the necessary degree of the polynomial function

    def __init__(self, degree):
        super(PolynomialModel, self).__init__() # the super() function is used to call a method from a parent class, in our case the nn.Module
        # Initialise the model parameters. These will be used as the parameters we update throughout the computations of the model.
        self.degree = degree # Here, we define the degree of the polynomial
        self.weight = nn.Parameter(torch.randn(degree + 1, requires_grad=True)) # Here, we initialise the weight parameter for the polynomial terms
        self.bias = nn.Parameter(torch.randn(1, requires_grad=True)) # Here, we initialise the weight parameter of the bias term

    #This defines the forward propagation steps of the model (common in all training instances)
    def forward(self, x):
        # This returns the polynomial function. For instance, in the case of a second degree polynomial, we retrieve: self.weight * x**2 + self.weight * x + self.bias
        return torch.sum(self.weight * x**torch.arange(self.degree + 1).float(), dim=1, keepdim=True) + self.bias
```

This class already looks somewhat complicated, but its composition is actually quite logical. Let's decrypt the individual steps:

1. class `PolynomialModel()`:
 - This is the class which is defined to create the model of interest
 - It contains the entire model structure as well as all the required parameters to initalise and update throughout the model training and evaluation process
2. subclass `nn.Module`:
 - This is the base class for neural network modules
 - It provides a set of essential functionalities that make it easier to design, train, and manage complex neural networks such as:
      - Parameter management:
         - It handles the management of learnable parameters (weights biases)
         - Parameters are automatically registered and tracked
      - Submodule Nesting:
         - Enables to create complex network architectures by nesting different layers and components within each other
      - Gradient Calculation:
         - It allows gradients to be automatically calculated for the parameters during the backward pass
      - Forward Pass definition:
         - The autograd engine automatically records the operations in the forward pass to compute gradients later
 - All these operational parts are stored as subclasses
      - As such, it should always be initialised with a `super(, self).__init__()` argument
3. Initialisation of subclass `super(PolynomialModel, self).__init__()`:
  - Generally, the `super()` function is used to call a method from a parent class
  - When a subclass is created, it can inherit attributes and methods from its parent class
  - In our case, the call is used to call the constructor of the `nn.Module` class (the parent class of PolynomialModel).
  - This ensures that the necessary initialization logic of nn.Module is executed, which is important for proper functioning of the PolynomialModel as a subclass of nn.Module
4. Initialisation of the `self.weight` and `self.bias` parameters:
  - `nn.Parameter()`:
    - Assigns the structure of the input tensor such that it can be handled by the `nn.Module` class
  - `torch.randn()`:
    - Here, we create a random tensor to initialise the respective parameter
    - This is a common approach in training ML models
  - `degree + 1`:
    - This is just necessary in a polynomial model where we define the random parameters to be different based on the degree
  - `requires_grad = True`:
    - This ensures that the gradients of the respective parameters are tracked
    - This is necessary to perform gradient descent
5. Initialisation of the `forward()` function:
  - Defines the forward propagation computation of the model
  - Every subclass of `nn.Module` needs to override the forward function

##### 1.2.3.2: Training Loop and Validation (including Early Stopping)

Let's now consider the training and validation loops.

In essence, we can distribute each training and validation loop into the following steps.

**1. Initialization**:

- Lists, train_losses, val_losses, and epochs are defined to store loss values and epoch counts.
- The total number of epochs is set to num_epochs = 10000.

**2. Training Loop**:

- Loop through each epoch.
 - Inside the loop:
    - Create input features for the polynomial model using the training data.
    - Set the model to training mode using model.train().
    - Perform forward propagation to obtain predictions (y_train_pred) on the training data.
    - Calculate the loss on the training data using the defined loss function (loss_train).
    - Reset gradients using optimizer.zero_grad().
    - Perform backward propagation to compute gradients (loss_train.backward()).
    - Update the model's parameters using the optimizer (optimizer.step()).

**3. Validation (Evaluation) Loop**:

 - Inside the same epoch loop:
    - Set the model to evaluation mode using model.eval().
    - Within a context of inference mode (with torch.inference_mode()):
    - Create input features for the polynomial model using the validation data.
    - Obtain predictions (y_val_pred) on the validation data.
    - Calculate the loss on the validation data using the defined loss function (loss_val).
    - Create an early stopping element which checks whether the calculated loss is lower than the optimal loss after an epoch. If after n iterations (n = patience parameter) the loss has not improved relative to the best evaluation loss, then stop the algorithm (early_stop_counter).

**4. Printing and Logging**:

 - Print progress updates every 100 epochs:
 - Display the epoch number, training set loss, and validation set loss.
 - Append the loss values and epoch to their respective lists for plotting
 later.

Although 1) and 4) are usually executed as well, they do not constitute the main parts of the actual training loop. The two most important parts are the actual training and validation loop.

**Early Stopping**

We should probably still explain the concept of early stopping a little better.

```python
# Define the loss values and the epoch counts
train_losses = []
val_losses = []
epochs = []
# Define the number of epochs
num_epochs = 10000
# Define early stopping parameters
patience = 300  # Number of epochs to wait for improvement
best_val_loss = float('inf')
early_stop_counter = 0

best_model_state = None  # Store the best model parameters

# Define the loop
for epoch in range(num_epochs):
  # Training loop
    # 1. Create Input Features for Polynomiasl model
    x_train_inputs = x_train.unsqueeze(1).pow(torch.arange(degree + 1).float())  # Create input features for the polynomial model
    # 2. Activate Training Mode of the model
    model.train()
    # 3. Forward Propagation of the model on the training data (using the forward() method)
    y_train_pred = model(x_train_inputs)
    # 4. Calculation of the Loss function on training data
    loss_train = criterion_loss(y_train_pred, y_train.unsqueeze(1))
    # 5. Backward Propagation of the Optimiser
    optimizer.zero_grad()
    # 6. Backward Propagation of the Loss
    loss_train.backward()
    # 7. Progress of the Optimiser
    optimizer.step()
  # Evaluation loop
    # 1. Activate the Validation / Test Mode of the model
    model.eval()
    with torch.inference_mode():
      # 2. Configure again the data
      x_val_inputs = x_val.unsqueeze(1).pow(torch.arange(degree + 1).float())  # Create input features for the polynomial model
      # 3. Assign model to the test data
      y_val_pred = model(x_val_inputs)
      # 4. Calculation of the Loss function on test data
      loss_val = criterion_loss(y_val_pred, y_val.unsqueeze(1))
      # 5. Early stopping check
      if loss_val < best_val_loss:
          best_val_loss = loss_val
          early_stop_counter = 0
          # Save the best model
          best_model_state = model.state_dict().copy()
      else:
          early_stop_counter += 1

    # Print an Epoch function:
    if epoch % 100 == 0:
      print('-'*100)
      print(f"Epoch: {epoch} | Training set loss: {loss_train} | Validation set loss: {loss_val}")
      print('-'*100)
      # Append the train, val losses throughout the epochs
      train_losses.append(loss_train)
      val_losses.append(loss_val)
      epochs.append(epoch)
    # Check for early stopping
      if early_stop_counter >= patience:
          print("Early stopping triggered.")
          break
```

Importantly, we can print the loss function. This function shows us whether the model training was successful and, as such, if our model improved its performance to approximate the underlying data structure of the sample data.

```python
# Plot the training and validation loss as function of the epochs
plt.plot(epochs, [i.item() for i in train_losses], label="Training Set loss")
plt.plot(epochs, val_losses, label="Evaluation Set loss")
plt.title("Training and Evaluation Set Loss curves")
plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.legend();
```

###### 1.2.3.2.1: Early Stopping

Let's explain the concept of early stopping a little more in detail.

Early stopping is a technique commonly used during the training of machine learning models to prevent overfitting and improve generalization. It involves monitoring the model's performance on a validation set and stopping the training process when the performance starts to degrade. The intuition behind early stopping is that as the model continues to train, it may start to fit noise in the training data, causing its performance on the validation set to worsen.

The main concepts of early stopping are the following:

- Monitoring Validation Loss:
   - During training, the model's performance is evaluated on a separate validation set that it has not seen during training. The validation loss, typically calculated using a loss function, is a measure of how well the model is generalizing to unseen data.

- Early Stopping Condition:
   - The training process is monitored over epochs. If the validation loss stops improving (or starts increasing) for a certain number of consecutive epochs, an early stopping condition is triggered.

- Preventing Overfitting:
   - When the model's performance on the validation set stops improving, it may indicate that the model is starting to overfit the training data. Overfitting occurs when the model becomes too complex and captures noise in the training data, leading to poor performance on unseen data.

- Terminating Training:
   - Once the early stopping condition is met, the training process is halted even if the specified number of epochs has not been completed. This prevents the model from further fitting noise in the training data.

Especially, Early Stopping is applied as it delivers the following benefits:

- Reduced Overfitting:
   - Early stopping helps prevent the model from becoming overly specialized to the training data and improves its ability to generalize to new, unseen data.

- Saves Time and Resources:
   - It can save computational resources and time by stopping the training process early when further improvements are unlikely.

- Simpler Models:
   - By stopping the training before the model becomes overly complex, early stopping often results in simpler and more interpretable models.

```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

## Part 1:

# Polynomial model class
class PolynomialModel(nn.Module): # The nn.Module: Is a subclass which contains all the building blocks to establish neural networks in PyTorch
    # Here, we initialise the class. Doing so, we need to use the:
      # - `self` argument
      # - `degree` argument: This will provide our model with the necessary degree of the polynomial function

    def __init__(self, degree):
        super(PolynomialModel, self).__init__() # the super() function is used to call a method from a parent class, in our case the nn.Module
        # Initialise the model parameters. These will be used as the parameters we update throughout the computations of the model.
        self.degree = degree # Here, we define the degree of the polynomial
        self.weight = nn.Parameter(torch.randn(degree + 1, requires_grad=True)) # Here, we initialise the weight parameter for the polynomial terms
        self.bias = nn.Parameter(torch.randn(1, requires_grad=True)) # Here, we initialise the weight parameter of the bias term

    #This defines the forward propagation steps of the model (common in all training instances)
    def forward(self, x):
        # This returns the polynomial function. For instance, in the case of a second degree polynomial, we retrieve: self.weight * x**2 + self.weight * x + self.bias
        return torch.sum(self.weight * x**torch.arange(self.degree + 1).float(), dim=1, keepdim=True) + self.bias

## Part 2:

# Define a list of learning rates to tune
learning_rates = [0.001, 0.01, 0.1, 1]

Best_Models = []
Best_Losses = []
# Loop over different learning rates
for lr in learning_rates:
  print(f"Training with learning rate: {lr}")
  # Create a PolynomialModel instance
  degree = 2  # Quadratic polynomial
  model = PolynomialModel(degree)

  # Define loss function and optimizer
  criterion_loss = nn.MSELoss()
  optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

  ## Part 3:

  # Define the loss values and the epoch counts
  train_losses = []
  val_losses = []
  epochs = []
  # Define the number of epochs
  num_epochs = 10000
  # Define early stopping parameters
  patience = 300  # Number of epochs to wait for improvement
  best_val_loss = float('inf')
  early_stop_counter = 0

  best_model_state = None  # Store the best model parameters

  # Define the loop
  for epoch in range(num_epochs):
    # Training loop
      # 1. Create Input Features for Polynomiasl model
      x_train_inputs = x_train.unsqueeze(1).pow(torch.arange(degree + 1).float())  # Create input features for the polynomial model
      # 2. Activate Training Mode of the model
      model.train()
      # 3. Forward Propagation of the model on the training data (using the forward() method)
      y_train_pred = model(x_train_inputs)
      # 4. Calculation of the Loss function on training data
      loss_train = criterion_loss(y_train_pred, y_train.unsqueeze(1))
      # 5. Backward Propagation of the Optimiser
      optimizer.zero_grad()
      # 6. Backward Propagation of the Loss
      loss_train.backward()
      # 7. Progress of the Optimiser
      optimizer.step()
    # Evaluation loop
      # 1. Activate the Validation / Test Mode of the model
      model.eval()
      with torch.inference_mode():
        # 2. Configure again the data
        x_val_inputs = x_val.unsqueeze(1).pow(torch.arange(degree + 1).float())  # Create input features for the polynomial model
        # 3. Assign model to the test data
        y_val_pred = model(x_val_inputs)
        # 4. Calculation of the Loss function on test data
        loss_val = criterion_loss(y_val_pred, y_val.unsqueeze(1))
        # 5. Early stopping check
        if loss_val < best_val_loss:
            best_val_loss = loss_val
            early_stop_counter = 0
            # Save the best model
            best_model_state = model.state_dict().copy()
        else:
            early_stop_counter += 1

      # Print an Epoch function:
      if epoch % 100 == 0:
        print('-'*100)
        print(f"Epoch: {epoch} | Training set loss: {loss_train} | Validation set loss: {loss_val}")
        print('-'*100)
        # Append the train, val losses throughout the epochs
        train_losses.append(loss_train)
        val_losses.append(loss_val)
        epochs.append(epoch)
      # Check for early stopping
        if early_stop_counter >= patience:
            print("Early stopping triggered.")
            break
  # Get a list of the best losses per LR parameter
  Best_Models.append(best_model_state)
  Best_Losses.append(best_val_loss)

## Part 4:

# Find the best model throughout all hyperparameter combinations
best_model_state_final = Best_Models[np.argmin(np.array(Best_Losses))]
# Load the best model parameters
model.load_state_dict(best_model_state_final)

# Apply the best model to the test data
model.eval()
with torch.no_grad():
    x_test_inputs = x_test.unsqueeze(1).pow(torch.arange(degree + 1).float())
    y_test_pred = model(x_test_inputs)
    test_loss = criterion_loss(y_test_pred, y_test.unsqueeze(1))

# Print the test loss
print('='*100)
print(f"Final Loss of the Test set: {test_loss}")
print('='*100)

# Generate predictions for the entire x range
inputs_all = x.unsqueeze(1).pow(torch.arange(degree + 1).float())  # Create input features for the entire x range
predictions_all = model(inputs_all)

# Generate predictions for the test set
inputs_test = x_test.unsqueeze(1).pow(torch.arange(degree + 1).float())  # Create input features for the test set
predictions_test = model(inputs_test)

# Plot the original polynomial and the approximated polynomial with colored data points
plt.figure(figsize=(10, 8))
plt.plot(x, y, label=f'Original Polynomial: y = {weight_2}*x^2 + {weight_1}*x + {bias}', color='lightblue')
plt.plot(x, predictions_all.detach(), label=f'Approximated Polynomial', color='lightgray')
plt.scatter(x_train, y_train, c='blue', label='Train Data')
plt.scatter(x_val, y_val, c='orange', label='Validation Data')
plt.scatter(x_test, y_test, c='green', label='Test Data')
plt.scatter(x_test, predictions_test.detach(), c='lightgreen', label='Predicted Test Data')
for i in range(len(x_test)):
    plt.plot([x_test[i], x_test[i]], [y_test[i], predictions_test[i].item()], linestyle='dotted', color='gray')
    plt.plot([x_test[i], x_test[i]], [y_test[i], predictions_test[i].item()], linestyle='dotted', color='gray')
    plt.plot([x_test[i], x_test[i]], [y_test[i], predictions_test[i].item()], linestyle='dotted', color='gray')
plt.title("Approximating Polynomial with PyTorch Model and Data Splits")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

#### 1.2.4: Model Depolyment and Saving

We have now a fully functioning process to create, run and test a model. Although we can easily run such a model multiple times, the computational complexity increases substantially with larger models. As such, we need to be able to store the model configuration which suits the data best in order for us to quickly load and apply the configuration, instead of re-calculating the model from scratch.

 You can use the `torch.save()` function to save the model's state dictionary (which contains all the learned parameters) to a file, and later use the `torch.load()` function to load the state dictionary back and assign it to the model. Here's how you can do it:

```python
# Create the path directory where you want to store the model:
model_dir = 'H:\00_Neural_Nets\Model\Model_Configuration'
file_name = 'best_model_state_final.pth'
# Save the best model's state dictionary to a file
torch.save(obj = best_model_state_final, f = model_dir + '\\' + file_name)

# Load the saved state dictionary and assign it to the model
loaded_state_dict = torch.load(f = model_dir + '\\' + file_name)
model.load_state_dict(loaded_state_dict)

# Apply the loaded model to the test data
model.eval()
with torch.no_grad():
    x_test_inputs = x_test.unsqueeze(1).pow(torch.arange(degree + 1).float())
    y_test_pred = model(x_test_inputs)
    test_loss = criterion_loss(y_test_pred, y_test.unsqueeze(1))

# Print the test loss
print('='*100)
print(f"Final Loss of the Test set: {test_loss}")
print('='*100)
```
